# Naive Bayes Classifier

```python
weather = ['Sunny', 'Sunny', 'Overcast', 'Rainy', 'Rainy', 'Rainy',
          'Overcast', 'Sunny', 'Sunny', 'Rainy', 'Sunny', 'Overcast',
          'Overcast', 'Rainy']

temp = ['Hot', 'Hot', 'Hot', 'Mild', 'Cool', 'Cool', 'Cool', 'Mild',
       'Cool', 'Mild', 'Mild', 'Mild', 'Hot', 'Mild']

play = ['No', 'No', 'Yes', 'Yes', 'Yes', 'No', 'Yes', 'No', 'Yes',
       'Yes', 'Yes', 'Yes', 'Yes', 'No']
```

```python
from sklearn import preprocessing

le = preprocessing.LabelEncoder()

weather_encoded = le.fit_transform(weather)

print(weather_encoded)
```

```python
temp_encoded = le.fit_transform(temp)

label = le.fit_transform(play)

print(temp_encoded)
print(label)
```

```python
features = list(zip(weather_encoded, temp_encoded))

features
```

```python
from sklearn.naive_bayes import GaussianNB

model = GaussianNB()

model.fit(features, label)

predicted = model.predict([[0,2]])
print("Predicted Value:", predicted)
```

# Decision Tree Classifier

```python
import pandas as pd

from sklearn.tree import DecisionTreeClassifier
```

```python
col_names = ['pregnant', 'glucose', 'bp', 'skin', 'insulin',
            'bmi', 'pedigree', 'age', 'label']

pima = pd.read_csv('https://raw.githubusercontent.com/ardhiraka/PFDS_sources/master/diabetes.csv', header=None, names=col_names)
```

```python
pima.head()
```

```python
pima.info()
```

```python
numer = ['pregnant', 'insulin', 'bmi', 'age', 'glucose', 'bp', 'pedigree', 'label']

for col in numer:
    pima[col] = pd.to_numeric(pima[col], errors='coerce')
```

```python
pima.dropna(inplace=True)
```

```python
feature_cols = ['pregnant', 'insulin', 'bmi', 'age', 'glucose', 'bp', 'pedigree']

X = pima[feature_cols]
y = pima.label
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

```python
clf = DecisionTreeClassifier()

clf = clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)
```

```python
metrics.accuracy_score(y_test, y_pred)
```

```python
pd.Series(y_pred).value_counts()
```

# Random Forest

```python
from sklearn import datasets

iris = datasets.load_iris()
```

```python
print(iris.target_names)

print(iris.feature_names)
```

```python
print(iris.data[0:5])
```

```python
#buat data frame

import pandas as pd

data=pd.DataFrame({'sepal length': iris.data[:,0],
                  'sepal width': iris.data[:,1],
                  'petal length': iris.data[:,2],
                  'petal width': iris.data[:,3],
                  'species': iris.target
})

data.head()
```

```python
from sklearn.model_selection import train_test_split

X=data[['sepal length', 'sepal width', 'petal length', 'petal width']]
y=data['species'] #label

#split dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3) #70% train dan 30% test
```

```python
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier(n_estimators=100)

clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)
```

```python
from sklearn import metrics

print("Accuracy:" , metrics.accuracy_score(y_test, y_pred))
```

```python
clf.predict([[3, 5, 4, 2]])
```

```python
feature_imp = pd.Series(clf.feature_importances_, index=iris.feature_names).sort_values(ascending=False)
feature_imp
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

sns.barplot(x=feature_imp, y=feature_imp.index)

plt.show()
```

```python
import numpy as np
```

```python
train = pd.read_csv('https://raw.githubusercontent.com/ardhiraka/PFDS_sources/master/Final_Dataset/train.csv')
```

```python
train.describe()
```

```python
print("Train Data:")
print(train.isnull().sum(), "\n")
```

```python
train.fillna(train.mean(), inplace=True)
train.isnull().sum()
```

```python
train.Gender.mode()
```

```python
train.Gender.fillna(train.Gender.mode()[0], inplace=True)
train.Married.fillna(train.Married.mode()[0], inplace=True)
train.Dependents.fillna(train.Dependents.mode()[0], inplace=True)
train.Self_Employed.fillna(train.Self_Employed.mode()[0], inplace=True)
train.isnull().sum()
```

```python
train.Loan_Amount_Term=np.log(train.Loan_Amount_Term)
```

```python
X = train.drop('Loan_Status',1)
y = train.Loan_Status
```

```python
X=pd.get_dummies(X)
train=pd.get_dummies(train)
```

```python
X.head()
```

```python
y.head()
```

```python
x_train, x_cv, y_train, y_cv = train_test_split(X, y, test_size=0.2)
```

```python
## Logistic Regression
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(x_train, y_train)
```

```python
pred_cv = model.predict(x_cv)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
print(accuracy_score(y_cv, pred_cv))
matrix=confusion_matrix(y_cv, pred_cv)
print(matrix)
```

```python
## Decision Tree

from sklearn import tree

dt = tree.DecisionTreeClassifier()
dt.fit(x_train, y_train)
```

```python
pred_cv1= dt.predict(x_cv)
```

```python
print(accuracy_score(y_cv, pred_cv1))
matrix1=confusion_matrix(y_cv, pred_cv1)
print(matrix1)
```

```python
## Random Forest

from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier()
rf.fit(x_train, y_train)
```

```python
pred_cv2= rf.predict(x_cv)

print(accuracy_score(y_cv, pred_cv2))
matrix2=confusion_matrix(y_cv, pred_cv2)
print(matrix2)
```

```python
## SVM

from sklearn import svm

svm_model = svm.SVC()
svm_model.fit(x_train, y_train)
```

```python
pred_cv3= svm_model.predict(x_cv)

print(accuracy_score(y_cv, pred_cv3))
matrix3=confusion_matrix(y_cv, pred_cv3)
print(matrix3)
```

```python
## Naive Bayes

from sklearn.naive_bayes import GaussianNB

nb = GaussianNB()
nb.fit(x_train, y_train)

pred_cv4= nb.predict(x_cv)

print(accuracy_score(y_cv, pred_cv4))
matrix4=confusion_matrix(y_cv, pred_cv4)
print(matrix4)
```

```python
## KNN

from sklearn.neighbors import KNeighborsClassifier

KNN = KNeighborsClassifier()
KNN.fit(x_train, y_train)

pred_cv5= KNN.predict(x_cv)

print(accuracy_score(y_cv, pred_cv5))
matrix5=confusion_matrix(y_cv, pred_cv5)
print(matrix5)
```
